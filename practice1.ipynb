```python
import xlrd
import xlwt

# 讀取士林.xls檔案
source_file = "士林.xls"
source_wb = xlrd.open_workbook(source_file)
source_ws = source_wb.sheet_by_index(0)  # 假設要從第1個工作表複製

# 讀取113年4月.xls檔案
target_file = "113年4月.xls"
target_wb = xlrd.open_workbook(target_file, formatting_info=True)
target_ws = target_wb.sheet_by_index(20)  # 第21個工作表的索引是20

# 創建一個新的Workbook來寫入數據
new_wb = xlwt.Workbook()
new_ws = new_wb.add_sheet(target_ws.name)  # 使用現有工作表名稱

# 複製資料
for row in range(7, 16):  # B8到T16範圍的行數 (xlrd從0開始索引)
    for col in range(1, 20):  # B到T範圍的列數 (xlrd從0開始索引)
        new_ws.write(row, col, source_ws.cell_value(row, col))

# 將新Workbook的數據寫入到原有的113年4月.xls檔案的第21個工作表
with open(target_file, 'wb') as f:
    new_wb.save(f)
```
